## (c)

##### AT Feature 

```python
plot_x=[]
dataSet_y_PE = data_frame['PE'].values
dataSet_x_AT = data_frame['AT'].values
regr_AT = LinearRegression().fit(dataSet_x_AT.reshape(-1, 1), dataSet_y_PE)
AT = sm.add_constant(dataSet_x_AT)
model_AT=sm.OLS(dataSet_y_PE,AT)
est_AT=model_AT.fit()
print('Linear regression score is',regr_AT.score(dataSet_x_AT.reshape(-1, 1), dataSet_y_PE))
print ('Interception is:',float(regr_AT.intercept_))
print('Coefficient is:',float(regr_AT.coef_[0]))
#print('Mse is:',metrics.mean_squared_error(dataSet_y_PE, regr_AT.predict(dataSet_x_AT.reshape(-1, 1))))
plt.figure(figsize=(12,10))
plt.scatter(dataSet_x_AT, dataSet_y_PE,  color='green')
plt.plot(dataSet_x_AT, regr_AT.predict(dataSet_x_AT.reshape(-1, 1)), color='red', linewidth=1)
plt.title('Linear regression between Temperature and Electrical Energy')
plt.xlabel('Temperature')
plt.ylabel('Electrical Energy')
plt.show()
print(est_AT.summary())
plot_x.append(float(regr_AT.coef_[0]))
```

```python
outlier_AT=[]
mse_at=[]
for i in range(len(dataSet_y_PE)):
    mse_at.append(abs((dataSet_y_PE[i]**2)-(est_AT.fittedvalues[i]**2)))
percent_1=np.percentile(stats.zscore(mse_at),1)
percent_99=np.percentile(stats.zscore(mse_at),99)
for index in range(len(mse_at)):
    if stats.zscore(mse_at)[index]<percent_5 or stats.zscore(mse_at)[index]>percent_95:
        outlier_AT.append(index)
print('The serial number of outliers are:',outlier_AT)
```

### interaction terms 

```python
# interaction terms
inter_all = np.column_stack((random_AT, random_V,random_AP,random_RH,random_AT*random_V,random_AT*random_AP,random_AT*random_RH,random_V*random_AP,random_V*random_RH,random_AP*random_RH,random_AT*random_V*random_AP,random_AT*random_V*random_RH,random_AT*random_AP*random_RH,random_V*random_AP*random_RH,random_AT*random_V*random_AP*random_RH))
inter_all= sm.add_constant(inter_all)
model_inter_all = sm.OLS(random_PE,inter_all).fit()
inter_params=['Interception','AT','V','AP','RH','AT*V','AT*AP','AT*RH','V*AP','V*RH','AP*RH','AT*V*AP','AT*V*RH','AT*AP*RH','V*AP*RH','AT*V*AP*RH']
inter_params_list=list(zip(inter_params,model_inter_all.params))
MSE_train=metrics.mean_squared_error(random_PE, model_inter_all.fittedvalues)
print('Interaction Model Train MSE is:',MSE_train)
inter_test=np.column_stack((remain_AT, remain_V,remain_AP,remain_RH,remain_AT*remain_V,remain_AT*remain_AP,remain_AT*remain_RH,remain_V*remain_AP,remain_V*remain_RH,remain_AP*remain_RH,remain_AT*remain_V*remain_AP,remain_AT*remain_V*remain_RH,remain_AT*remain_AP*remain_RH,remain_V*remain_AP*remain_RH,remain_AT*remain_V*remain_AP*remain_RH))
inter_test= sm.add_constant(inter_test)
MSE_test=metrics.mean_squared_error(remain_PE, model_inter_all.predict(inter_test))
print('Interaction Model Test MSE is:',MSE_test)
print(inter_params_list)
print(model_inter_all.summary())
```

## (i)KNN Contrast

```python
#normalize features
def Normalize(data):
    m = np.mean(data)
    mx = max(data)
    mn = min(data)
    return [(float(i) - m) / (mx - mn) for i in data]
No_AT=(np.array(Normalize(new_data['AT'].values))).reshape(-1,1)
No_V=(np.array(Normalize(new_data['V'].values))).reshape(-1,1)
No_AP=(np.array(Normalize(new_data['AP'].values))).reshape(-1,1)
No_RH=(np.array(Normalize(new_data['RH'].values))).reshape(-1,1)
No_PE=np.array(Normalize(new_data['PE'].values))
knn_train=np.hstack((No_AT,No_V,No_AP,No_RH))
No_AT_remain=(np.array(Normalize(data_remain['AT'].values))).reshape(-1,1)
No_V_remain=(np.array(Normalize(data_remain['V'].values))).reshape(-1,1)
No_AP_remain=(np.array(Normalize(data_remain['AP'].values))).reshape(-1,1)
No_RH_remain=(np.array(Normalize(data_remain['RH'].values))).reshape(-1,1)
No_PE_remain=np.array(Normalize(data_remain['PE'].values))
knn_train_remain=np.hstack((No_AT_remain,No_V_remain,No_AP_remain,No_RH_remain))
normal_knn_train_errors=[]
nor_train_error={}
normal_knn_test_errors=[]
nor_test_error={}
for i in range(1,101,1):    
    knn_E = KNeighborsRegressor(n_neighbors=i)
    knn_E.fit(knn_train,No_PE) 
    score_train_normalzie=knn_E.score(knn_train,No_PE)
    score_test_normalzie=knn_E.score(knn_train_remain,No_PE_remain)
    normal_knn_train_errors.append(1-score_train_normalzie)
    normal_knn_test_errors.append(1-score_test_normalzie)
    nor_train_error[i]=score_train_normalzie
    nor_test_error[i]=score_test_normalzie
normal_k_train_better=max(nor_train_error,key=nor_train_error.get)
normal_k_test_better=max(nor_test_error,key=nor_test_error.get)
print('For KNN Regression(Normalize) useing train data,its best k is:',normal_k_train_better,',best error rate is',min(normal_knn_train_errors))
print('For KNN Regression(Normalize) useing test data,its best k is:',normal_k_test_better,',best error rate is',min(normal_knn_test_errors))
plt.figure(figsize=(18,10))
plt.plot((1/np.array(range(1,101,1))),normal_knn_train_errors,c='red',marker='.',alpha=1,label='Normalize features train errors rate')
plt.plot((1/np.array(range(1,101,1))),normal_knn_test_errors,c='green',marker='.',alpha=1,label='Normalize features test errors rate')
plt.title("KNN Regression Using Normalize Features")
plt.ylabel("Error Rate")
plt.xlabel("1/k")
plt.legend(loc='upper right')
plt.show()
```
