### (iii)

Stratified cross validation belongs to the cross-validation type. Stratified means that the proportional relationship of each category in the original data is maintained in each fold.Because the numbers of bending are much less than others, we should use tratified cross validation to avoid class imbalance.

For the cross validation, the right way is to use it at the beginning. By compared, the wrong way is to use cross validation after training.

In this question,I will use REFCV to prune the features. So that,there is no need to calculate p-value,since all these processes have been finished among the module.

```python
#1 for bendings;0 for others
l_accurancy={}
for l in range(1,21):
#for l in range(5,6):
    instance_features=pd.DataFrame(columns=range(1,18*l+2))
    for ins in range(69):
        if ins<43:
            data_frame=train_data.iloc[480*ins:480*(ins+1)]
        elif i==43:
            data_frame=train_data.iloc[480*ins:480*(ins+1)-1]
        else:
            data_frame=train_data.iloc[480*ins-1:480*(ins+1)-1]        
        data_frame = data_frame.reset_index(drop = True)
        feature_list=[]
        m=int(data_frame.shape[0]/l)
        for k in range(1,l+1):
            if k==l:
                data_print=data_frame[m*(k-1):data_frame.shape[0]]
            else:
                data_print=data_frame[m*(k-1):m*k]
            for n in range(1,7):
                data_print[str(n)] = data_print[str(n)].map(lambda x:float(x))
                feature_list.append(min(data_print[str(n)]))
                feature_list.append(max(data_print[str(n)]))
                feature_list.append(round(np.mean(data_print[str(n)]),2))
        feature_list.append(int(min(data_print['7'])))
        instance_features.loc[instance_features.shape[0]+1]=feature_list
    #print(instance_features)
    skf = StratifiedKFold(n_splits=5)
    score_total=0
    for train_index, test_index in skf.split(instance_features.iloc[:,:18*l], instance_features[18*l+1]):
        traindataset=instance_features.iloc[train_index.tolist(),]
        testdataset=instance_features.iloc[test_index.tolist(),]
        log_model = LogisticRegression()
        selector = RFECV(log_model)
        selector = selector.fit(traindataset.iloc[:,:18*l], traindataset[18*l+1])
        rankset=selector.ranking_
        goodlist=[]
        for i in range(len(rankset)):
            if rankset[i]==1:
                goodlist.append(i+1)
        goodlist.append(18*l+1)
        real_train=traindataset[goodlist]
        real_train_label=real_train[18*l+1]
        real_train_data=real_train.drop(18*l+1,axis=1)
        real_test=testdataset[goodlist]
        real_test_label=real_test[18*l+1]
        real_test_data=real_test.drop(18*l+1,axis=1)
        clf = LogisticRegression().fit(real_train_data, real_train_label)
        score_total=score_total+clf.score(real_test_data, real_test_label)
    l_accurancy[l]=score_total/5
#print(l_accurancy)
l_better=max(l_accurancy,key=l_accurancy.get)
```

```python
print('By the cross validation,the best l is:',l_better)
print('The best score is:',l_accurancy[l_better])
```

```python
gua_Bayes_accurancy={}
mul_Bayes_accurancy={}
for l in range(1,21):
    instance_features=pd.DataFrame(columns=range(1,18*l+2))
    for ins in range(69):
        if ins<43:
            data_frame=multi_train_data_norm.iloc[480*ins:480*(ins+1)]
        elif i==43:
            data_frame=multi_train_data_norm.iloc[480*ins:480*(ins+1)-1]
        else:
            data_frame=multi_train_data_norm.iloc[480*ins-1:480*(ins+1)-1]        
        data_frame = data_frame.reset_index(drop = True)
        feature_list=[]
        m=int(data_frame.shape[0]/l)
        for k in range(1,l+1):
            if k==l:
                data_print=data_frame[m*(k-1):data_frame.shape[0]]
            else:
                data_print=data_frame[m*(k-1):m*k]
            for n in range(1,7):
                data_print[str(n)] = data_print[str(n)].map(lambda x:float(x))
                feature_list.append(min(data_print[str(n)]))
                feature_list.append(max(data_print[str(n)]))
                feature_list.append(round(np.mean(data_print[str(n)]),2))
        feature_list.append(int(min(data_print['7'])))
        instance_features.loc[instance_features.shape[0]+1]=feature_list
    skf = StratifiedKFold(n_splits=5)
    gua_Bayes_score_total=0
    mul_Bayes_score_total=0
    for train_index, test_index in skf.split(instance_features.iloc[:,:18*l], instance_features[18*l+1]):
        traindataset=instance_features.iloc[train_index.tolist(),]
        testdataset=instance_features.iloc[test_index.tolist(),]
        
        #GaussianNB
        gua_Bayes = GaussianNB().fit(traindataset.iloc[:,:18*l], traindataset[18*l+1])
        gua_Bayes_score_total=gua_Bayes_score_total+gua_Bayes.score(testdataset.iloc[:,:18*l], testdataset[18*l+1])
        
        #MultinomialNB
        mul_Bayes = MultinomialNB().fit(traindataset.iloc[:,:18*l], traindataset[18*l+1])
        mul_Bayes_score_total=mul_Bayes_score_total+mul_Bayes.score(testdataset.iloc[:,:18*l], testdataset[18*l+1])
    #print(l)
    #print(gua_Bayes_score_total/5)
    #print(mul_Bayes_score_total/5)
    gua_Bayes_accurancy[l]=gua_Bayes_score_total/5
    mul_Bayes_accurancy[l]=mul_Bayes_score_total/5  
```
