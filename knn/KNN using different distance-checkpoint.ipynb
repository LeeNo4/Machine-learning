## (c)Classification using KNN on Vertebral Column Data Set

### Part i KNN code 

```python
def KNNclassify(testData,trainData,labels,k):
        dist = []
        for i in range(len(trainData)):
            td = trainData[i,:]   #train
            dist.append(np.linalg.norm(testData - td))   #Euclidean
        dist_collection = np.array(dist)
        dist_index = dist_collection.argsort()[0:k] 
        k_labels = labels[dist_index] 
        k_labels = list(k_labels)
        labels_count = {}
        for i in k_labels:
            labels_count[i] = k_labels.count(i) 
        testData_label = max(labels_count, key=labels_count.get)
        return int(testData_label)   
```

### Using self-finished knn 

```python
#Using self-finished knn
#the increments of k is -1 instead of -3
testData=data_test[:,0:data_test.shape[1]-1]
testDataLabels=data_test[:,data_test.shape[1]-1]
```

```python
#Because there are lots of loops, the running-time is pretty long.
test_accurancy={}#for all ks,the prediction accurancy
test_errors_table=[]
train_errors_table=[]
for k in range (208,0,-1):
    testData_label=[]#for one k,all prediction labels(testdataset)
    train_testData_label=[]#for one k,all prediction labels(traindataset)
    num=0.00#in the prediction array,the right numbers
    num2=0
    accurancy=0.00
    for m in range(len(testData)):
        testd=testData[m,:]
        #testData_label[num+1] = KNNclassify(testd,trainData,trainLabels,k)
        testData_label.append(KNNclassify(testd,trainData,trainLabels,k))
    for i in range(len(testData_label)):
        if(testData_label[i]==testDataLabels[i]):
            num=num+1
    for z in range(len(trainData)):
        testd=trainData[z,:]
        #testData_label[num+1] = KNNclassify(testd,trainData,trainLabels,k)
        train_testData_label.append(KNNclassify(testd,trainData,trainLabels,k))
    for y in range(len(trainData)):
        if(train_testData_label[y]==trainLabels[y]):
            num2=num2+1
    accurancy=num/(len(testData_label))
    test_errors_table.append(((len(testData))-num))
    train_errors_table.append(((len(trainData))-num2))
    test_accurancy[k]=accurancy
k_better=max(test_accurancy,key=test_accurancy.get)
error_knn_rate=1-test_accurancy[k_better]
best_errors_rate_for_all.append(error_knn_rate)
print('For self-code knn, its best k is:',k_better)
print('The best error rate is:',round(error_knn_rate,2))
```

```python
#Plot train and test errors
plt.figure(figsize=(18,10))
a=np.array(range(208,0,-1))
plt.plot(a,test_errors_table,c='red',marker='.',alpha=1,label='test errors rate')
plt.plot(a,train_errors_table,c='green',marker='.',alpha=1,label='train errors rate')
plt.title("KNN train and test errors")
plt.ylabel("errors numbers interm of k")
plt.xlabel("k values")
plt.legend(loc='lower right')
plt.show()
```

```python
#trainData-->trainLabels
#testData-->testDataLabels
#A
score_Manhattan={}
score_1=[]
error_rate=[]
metrics_error_table=[]
for i in range(1,201,5):    
    knn_Manhattan = KNeighborsClassifier(n_neighbors=i,p=1)#p=1 becomes Manhattan Distance
    knn_Manhattan.fit(trainData,trainLabels) 
    knn_Manhattan_predict_labels = knn_Manhattan.predict(testData)
    score=knn_Manhattan.score(testData,testDataLabels,sample_weight=None)
    score_Manhattan[i]=score
    score_1.append(score)
    error_rate.append(1-score)
plt.figure(figsize=(18,10))
plt.plot(np.array(range(1,201,5)),error_rate,c='red',marker='.',alpha=1)
k_Manhattan_better=max(score_Manhattan,key=score_Manhattan.get)
best_errors_rate_for_all.append(round(score_Manhattan[k_Manhattan_better],2))
plt.title("KNN(Manhattan Distance)")
plt.ylabel("Error Rate")
plt.xlabel("k values")
plt.show()
metrics_error_table.append(round(1-score_Manhattan[k_Manhattan_better],2))
print('By using Manhattan distance, the best k is:',k_Manhattan_better)
print('The best error rate is:',round(1-score_Manhattan[k_Manhattan_better],2))
```

## (e)Use weighted voting

```python
Weight_score_Euclidean={}
Weight_score_Manhattan={}
Weight_score_Chebyshev={}

Weight_score_Euclidean_p=[]
Weight_score_Manhattan_p=[]
Weight_score_Chebyshev_p=[]

Weight_error_Euclidean_rate=[]
Weight_error_Manhattan_rate=[]
Weight_error_Chebyshev_rate=[]
for i in range(1,201,5):    
    Weight_knn_Euclidean = KNeighborsClassifier(n_neighbors=i,weights='distance')
    Weight_knn_Manhattan = KNeighborsClassifier(n_neighbors=i,weights='distance',p=1)
    Weight_knn_Chebyshev = KNeighborsClassifier(n_neighbors=i,weights='distance',metric='chebyshev')
    
    Weight_knn_Euclidean.fit(trainData,trainLabels) 
    Weight_knn_Manhattan.fit(trainData,trainLabels) 
    Weight_knn_Chebyshev.fit(trainData,trainLabels) 
    
    knn_Euclidean_predict_labels = knn_Chebyshev.predict(testData)
    knn_Manhattan_predict_labels = knn_Chebyshev.predict(testData)
    knn_Chebyshev_predict_labels = knn_Chebyshev.predict(testData)
    
    score_1=Weight_knn_Euclidean.score(testData,testDataLabels,sample_weight=None)
    score_2=Weight_knn_Manhattan.score(testData,testDataLabels,sample_weight=None)
    score_3=Weight_knn_Chebyshev.score(testData,testDataLabels,sample_weight=None)
    
    Weight_score_Euclidean[i]=score_1
    Weight_score_Manhattan[i]=score_2
    Weight_score_Chebyshev[i]=score_3
    
    Weight_score_Euclidean_p.append(score_1)
    Weight_score_Manhattan_p.append(score_2)
    Weight_score_Chebyshev_p.append(score_3)
    
    Weight_error_Euclidean_rate.append(1-score_1)
    Weight_error_Manhattan_rate.append(1-score_2)
    Weight_error_Chebyshev_rate.append(1-score_3)

plt.figure(figsize=(18,10))
a=np.array(range(1,201,5))

plt.plot(a,Weight_error_Euclidean_rate,c='green',marker='.',alpha=1,label='Euclidean distance')
plt.plot(a,Weight_error_Manhattan_rate,c='red',marker='.',alpha=1,label='Manhattan distance')
plt.plot(a,Weight_error_Chebyshev_rate,c='yellow',marker='.',alpha=1,label='Chebyshev distance')

Weight_Euclidean_better=max(Weight_score_Euclidean,key=Weight_score_Euclidean.get)
Weight_Manhattan_better=max(Weight_score_Manhattan,key=Weight_score_Manhattan.get)
Weight_Chebyshev_better=max(Weight_score_Chebyshev,key=Weight_score_Chebyshev.get)

best_errors_rate_for_all.append(round(1-Weight_score_Euclidean[Weight_Euclidean_better],2))
best_errors_rate_for_all.append(round(1-Weight_score_Manhattan[Weight_Manhattan_better],2))
best_errors_rate_for_all.append(round(1-Weight_score_Chebyshev[Weight_Chebyshev_better],2))

plt.title("KNN voted by distance in different matrix")
plt.ylabel("Error Rate")
plt.xlabel("k values")
plt.legend(loc='lower right')
plt.show()
print('When we are using weighted decision')
print('The best k value for Euclidean KNN by distance weight is:',Weight_Euclidean_better,',the best error rate is:',round(1-Weight_score_Euclidean[Weight_Euclidean_better],3))
print('The best k value for Manhattan KNN by distance weight is:',Weight_Manhattan_better,',the best error rate is:',round(1-Weight_score_Manhattan[Weight_Manhattan_better],3))
print('The best k value for Chebyshev KNN by distance weight is:',Weight_Chebyshev_better,',the best error rate is:',round(1-Weight_score_Chebyshev[Weight_Chebyshev_better],3))
```
